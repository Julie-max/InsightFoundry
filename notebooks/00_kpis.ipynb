```python
import pandas as pd
import numpy as np
```

```python
# Update filename if needed
file_path = "../data/sales.csv"
df = pd.read_csv(file_path, encoding="latin1")

print("Shape:", df.shape)
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
# Example column names: quantity, unit_price, order_date, product_name, region

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['revenue'] = df['Sales']

print("Total Revenue:", df['revenue'].sum())
```

```python
monthly_revenue = (
    df.groupby(pd.Grouper(key='Order Date', freq='M'))['revenue']
      .sum()
      .reset_index()
)

monthly_revenue['pct_growth'] = monthly_revenue['revenue'].pct_change() * 100
monthly_revenue.head()
```

```python
top_products = (
    df.groupby('Product Name')['revenue']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products
```

```python
revenue_by_region = (
    df.groupby('Region')['revenue']
      .sum()
      .sort_values(ascending=False)
)

revenue_by_region
```

```python
monthly_profit = (
    df.groupby(pd.Grouper(key='Order Date', freq='M'))['Profit']
      .sum()
      .reset_index()
)
```

```python
monthly_combined = monthly_revenue.merge(
    monthly_profit,
    on='Order Date'
)

monthly_combined['profit_margin_pct'] = (
    monthly_combined['Profit'] /
    monthly_combined['revenue']
) * 100

monthly_combined.head()
```
