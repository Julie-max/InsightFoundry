```python
import pandas as pd
import numpy as np
```

```python
# Update filename if needed
file_path = "../data/sales.csv"
df = pd.read_csv(file_path)

print("Shape:", df.shape)
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
# Example column names: quantity, unit_price, order_date, product_name, region

df['order_date'] = pd.to_datetime(df['order_date'])
df['revenue'] = df['quantity'] * df['unit_price']

print("Total Revenue:", df['revenue'].sum())
```

```python
monthly_revenue = (
    df.groupby(pd.Grouper(key='order_date', freq='M'))['revenue']
      .sum()
      .reset_index()
)

monthly_revenue['pct_growth'] = monthly_revenue['revenue'].pct_change() * 100
monthly_revenue.head()
```

```python
top_products = (
    df.groupby('product_name')['revenue']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products
```

```python
revenue_by_region = (
    df.groupby('region')['revenue']
      .sum()
      .sort_values(ascending=False)
)

revenue_by_region
```
